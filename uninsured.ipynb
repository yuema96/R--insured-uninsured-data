**Part 1:** Calculate and plot the percentage of uninsured individuals based on data in `uninsured2012.csv`; your visualization should convey the percentage of uninsured individuals for each state.

```r
unin.csv<-read.csv(file ='Data\\uninsured2012.csv' )
```

```r
population <- aggregate(Population ~ State, unin.csv, sum)    
```

```r
population
```

```r
uninsured <- aggregate(Uninsured ~ State, unin.csv, sum) 
```

```r
uninsured$Uninsured/population$Population
```

```r
uninsured$State
```

```r
barplot(uninsured$Uninsured/population$Population*100~ uninsured$State,
        main='Percentage of Uninsured Individuals for Each State',
        xlab='State', 
        ylab='Percentage',
        col='yellow',
        ylim=c(0,25),
        las = 3,
        cex.names = 0.5)
```

**Part 2:** Sort the states by population, where the most populated state is listed first. Again plot the percentage of uninsured individuals for each state.

```r
final <- data.frame(population$State, population$Population,uninsured$Uninsured, uninsured$Uninsured/ population$Population )
```

```r
library(dplyr)
```

```r
p_sorted<-final %>%
   arrange(desc(final$population.Population),final$population.State, final$uninsured.Uninsured, final$uninsured.Uninsured.population.Population  )
```

```r
colnames(p_sorted) <- c("S_State","S_Population","S_Uninsured", "S_Percentage")
p_sorted
```

```r
 p_sorted$S_State<-factor(p_sorted$S_State, levels = p_sorted$S_State) #This is used to avoid automatically sorting by alphabetic in x axis in R plotting 
```

```r
barplot(p_sorted$S_Percentage*100~p_sorted$S_State,
        main='Percentage of Uninsured Individuals for Each State \n by Descending Order of Population',
        xlab='State', 
        ylab='Percentage',
        col='green',
        ylim=c(0,25),
        las = 3,
        cex.names = 0.5)
```
